## Propose a comprehensive data cleaning plan (feb 1-10)

### Subtask:
Outline detailed technical steps using Python code for cleaning and preparing the combined data from Feb 1st to Feb 10th for reporting.


```python
df_combined = pd.concat([dataset_1a, dataset_1b], ignore_index=True)
df_combined['transaction_date'] = pd.to_datetime(df_combined['transaction_date'])
df_combined = df_combined.sort_values(by=['item_id', 'shop_id', 'transaction_date']).reset_index(drop=True)
display(df_combined.head())
display(df_combined.info())
```

```python
df_combined['currency'] = df_combined['currency'].str.upper()

# Identify price anomalies based on the previous EDA
price_anomalies_mask = df_combined['price'].apply(lambda x: x not in [55000, 54000, 53000, 52000, 51000])

# Impute price anomalies with the mean price for the item and shop
df_combined['price'] = df_combined.groupby(['item_id', 'shop_id'])['price'].transform(lambda x: x.mask(price_anomalies_mask, x[~price_anomalies_mask].mean()))

display(df_combined.head())
display(df_combined.info())
```

```python
df_combined['daily_sold_change'] = df_combined.groupby(['item_id', 'shop_id'])['cumsum_sold_item'].diff().fillna(0)
df_combined['daily_stock_change'] = df_combined.groupby(['item_id', 'shop_id'])['remaining_stock'].diff().fillna(0)

# Identify anomalies in daily changes (negative sold change or positive stock change)
anomalies_mask = (df_combined['daily_sold_change'] < 0) | (df_combined['daily_stock_change'] > 0)

# Handle anomalies by setting daily changes to 0 (assuming these are errors and no sales/stock changes occurred)
df_combined.loc[anomalies_mask, ['daily_sold_change', 'daily_stock_change']] = 0

display(df_combined.head())
display(df_combined.info())
```

## Propose limited-time data analysis

### Subtask:
Suggest the most critical analysis to perform on the data from the second scenario if time is very limited.


```python
# Calculate metrics from the sample data (dataset_2b_sample) grouped by load_date and channel
sample_metrics = dataset_2b_sample.groupby(['load_date', 'channel']).agg(
    total_post_sample=('post_id', 'count'),
    total_likes_sample=('likes', 'sum'),
    total_views_sample=('views', 'sum')
).reset_index()

# Calculate avg_rate_likes_in_percentage from the sample data
# Assuming avg_rate_likes_in_percentage is total likes / total views
sample_metrics['avg_rate_likes_in_percentage_sample'] = (sample_metrics['total_likes_sample'] / sample_metrics['total_views_sample']).fillna(0)

# Merge with the report data (dataset_2a_report)
comparison_df = pd.merge(dataset_2a_report, sample_metrics, on=['load_date', 'channel'], how='left')

# Compare the metrics - Note: total_revenue cannot be calculated from sample data
print("Comparison of Report and Sample Metrics:")
display(comparison_df[['load_date', 'channel', 'total_post', 'total_post_sample', 'avg_rate_likes_in_percentage', 'avg_rate_likes_in_percentage_sample']])

# Identify significant discrepancies (e.g., more than 5% difference for total_post)
comparison_df['total_post_diff_percentage'] = abs(comparison_df['total_post'] - comparison_df['total_post_sample']) / comparison_df['total_post']
discrepancies = comparison_df[comparison_df['total_post_diff_percentage'] > 0.05]

print("\nRows with significant discrepancies in total_post (> 5%):")
display(discrepancies)

# Further analysis would involve investigating the source of these discrepancies.
```
